```python
import tqdm
from infections import get_infectable, InfectableType
from runtime_utils import *
import matplotlib.pyplot as plt
from IPython.display import clear_output
import seaborn as sns
sns.set(font_scale=2)

context = initialize()
context.persons[0].get_infected(get_infectable(InfectableType(3))) #zero patient
infected = []
hospitalized = []
dead = []
recovered = []

every = 10
for day in tqdm.tqdm(range(51)):
    simulate_day(context)
    infected.append(context.health_dept.situation["infected"])
    hospitalized.append(context.health_dept.situation["hospitalized"])
    dead.append(context.health_dept.situation["dead"])
    recovered.append(context.health_dept.situation["recovered"])
    if day % every == 0:
        clear_output()
        plt.figure(figsize=(20, 10))
        plt.plot(infected, label="infected")
        plt.plot(hospitalized, label="hospitalized")
        plt.plot(dead, label="dead")
        plt.plot(recovered, label="recovered")
        plt.legend()
        plt.show()
```
